```python
from aleatory.processes import BrownianMotion
brownian = BrownianMotion()
brownian.draw(n=100, N=100, colormap="cool", figsize=(8,6));
```

So I think this differs from arithmetic brownian motion above. This represents more of the path of stocks that start at zero evaluation. Baked into the SDE is the notion that higher valued stocks are more volatile and with higher drift parameter.

**Drift Parameter**?

This drift parameter most certainly exists and is correlated with *certain* factors such as hard work, etc. Although defining this drift parameter is rather difficult and our metrics fails us when we describe "success" when it inherently lacks a numerical metric. This drift parameter only increases the likelihood of an upward trajectory without guarantee. While starting out at a higher position (in economic model this is financial asset) means your drift parameter doesn't have to be wickedly large, just not too negative, in order to be deemed "high" in the eyes of others.
