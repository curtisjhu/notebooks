## Runge-Kutta example

$$ \frac{dy}{dt} = sin^2(t) y $$

Why this not working? Not runge-kutta

```python
# dx/dt = x(1-x)

starting_pos = np.linspace(0, 2.4, 12)
for init in starting_pos:

	values = np.array([])
	times = np.linspace(0, 100, 100)
	x = init
	for t in times:
		deriv = x * (1 - x)
		x += deriv
		values = np.append(values, x)
	plt.plot(times, values)
plt.show()
```
