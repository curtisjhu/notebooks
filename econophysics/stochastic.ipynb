Brownian bridge is just a Wiener process such that the ends are contrained to finite values:

$$ B_t = (W_t | W_0 = a, W_T = b), t \in [0, T] $$

```python
from aleatory.processes import BrownianBridge

bb = BrownianBridge(end=1)
bb.draw(n=50, N=500);
```
