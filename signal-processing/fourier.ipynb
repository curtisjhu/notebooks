```python
import matplotlib.pyplot as plt
import numpy as np
```

```python

x = np.linspace(-10, 10, 100)
W = np.array([1, 2, 3, 4])

y = 1/np.linalg.norm(W) * np.exp(- np.pi * x**2/ np.linalg.norm(W)**2)

plt.plot(x, y)
plt.xlabel("k space")
plt.ylabel("H(k)")
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate data
x = np.linspace(-10, 100, 100)
y = np.linspace(90, 110, 100)
X, Y = np.meshgrid(x, y)
Z = np.abs(X) + np.abs(Y)
Z2 = np.square(X) + np.square(Y)
Z3 = 100 - 2*x

# Create contour plot
plt.contour(X, Y, Z)
plt.contour(X, Y, Z2)
plt.plot(x, Z3)
plt.colorbar()  # Add a colorbar
plt.title('Contour Plot')
plt.xlabel('X')
plt.ylabel('Y')
plt.show()
```

```python
def U(A):
	Z = A/2
	return 15*A - 15*np.power(A, 2/3) - 25.5e-11 * Z*Z/(8.85e-12 * 3.14159 * np.power(A, 2/3))
```

```python
A = np.linspace(2, 300)
U_y = np.array([U(a) for a in A])

plt.plot(A, U_y)
plt.show()
```

```python

A = np.linspace(2, 300, 300-2)
b = np.array([U(a)/a for a in A])

plt.plot(A, b)
plt.show()

for a, b_ in zip(A, b):
	print(a, b_)
```

```python


M = 50
m = np.arange(-M, M)
dx = 1e-1
k = np.linspace(-3/dx, 3/dx)

m, k = np.meshgrid(m, k)

S = dx * 1/(1-np.exp(-2j * np.pi*k*m*dx))


fig, ax = plt.subplots(subplot_kw={"projection": "3d"})
ax.plot_surface(k, m, S)
ax.set_xlabel("m")
ax.set_ylabel("k")
```
