# Activations

```python
import keras
import matplotlib.pyplot as plt
import numpy as np
```

## Linear

```python
x = np.arange(-3, 3, 0.1)
y = keras.activations.linear(x)

plt.plot(x, y)
plt.show()
```

## RELU

```python
x = np.arange(-3, 3, 0.1)
y = keras.activations.relu(x)

plt.plot(x, y)
plt.show()
```

## Leaky ReLu

```python
x = np.arange(-3, 3, 0.1)
y = keras.activations.leaky_relu(x)

plt.plot(x, y)
plt.show()
```

## Sigmoid

$$ \frac{1}{1 + e^{-x}} $$

```python
x = np.arange(-8, 8, 0.1)
y = keras.activations.sigmoid(x)

plt.plot(x, y)
plt.show()
```

## Tanh

$$ tanh(x) = \frac{sinh(x)}{cosh(x)} = \frac{e^x - e^{-x}}{e^x + e^{-x}} $$

```python
x = np.arange(-8, 8, 0.1)
y = keras.activations.tanh(x)

plt.plot(x, y)
plt.show()
```

## Softmax

$$ \begin{bmatrix} x_1 \\ x_2 \\ ... \\ x_n \end{bmatrix} \rightarrow x_j \rightarrow \frac{e^{x_j}}{\sum_j e^{x_j}} $$

```python
x = np.arange(0, 10, 0.1)
y = keras.activations.softmax(x)

plt.plot(x, y)
plt.show()
```

## SeLU

$$ \begin{cases} k \cdot x \quad \text{if} \quad x > 0 \\ k \cdot \alpha \cdot e^x - 1  \end{cases} $$ 

$\alpha = 1.67326324$ and $k = 1.05070098$

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.selu(x)

plt.plot(x, y)
plt.show()
```

## Silu

$$ x * \sigma(x)$$

```python

x = np.arange(-10, 10, 0.1)
y = keras.activations.silu(x)
# y = x * keras.activations.sigmoid(x)

plt.plot(x, y)
plt.show()
```

## Softplus

$$ \log{(e^x + 1)} $$

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.softplus(x)

plt.plot(x, y)
plt.show()
```

## Softsign

$$ \frac{x}{|x| + 1} $$

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.softsign(x)

plt.plot(x, y)
plt.show()
```

## Mish

$$ x * \tanh(softplus(x)) $$

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.mish(x)

plt.plot(x, y)
plt.show()
```

## ELu

$$ \begin{cases} x \quad \text{if} \quad x > 0 \\ \alpha \cdot e^x - 1  \end{cases} $$

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.elu(x)

plt.plot(x, y)
plt.show()
```

## GELu

$$ gelu(x) = x \cdot P(X \leq x) \quad , \quad P(X) \sim N(0, 1)$$

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.gelu(x)

plt.plot(x, y)
plt.show()
```

## Hard Silu

```python

x = np.arange(-10, 10, 0.1)
y = keras.activations.hard_silu(x)

plt.plot(x, y)
plt.show()
```

## Hard Sigmoid

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.hard_sigmoid(x)

plt.plot(x, y)
plt.show()
```

## Exponential

```python
x = np.arange(-10, 10, 0.1)
y = keras.activations.exponential(x)

plt.plot(x, y)
plt.show()
```
