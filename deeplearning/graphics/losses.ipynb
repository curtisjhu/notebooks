```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def MSE(y, y_actual):
	# L2 Loss
	return 0.2 * (y-y_actual)**2
def MAE(y, y_actual):
	# L1 Loss
	return np.abs(y-y_actual)
def SMAE(y, y_actual):
	# Huber Loss
	# MSE when close
	delta = 2
	diff = y - y_actual
	for i, d in enumerate(diff):
		if np.abs(d) <= delta:
			diff[i] = 0.5 * d**2
		else:
			diff[i] = delta * np.abs(d) - 0.5 * delta**2
	return diff

def log_cosh(y, y_actual):
	return np.log(np.cosh(y - y_actual))

LEN = 40
y_actual = np.linspace(-3, 3, LEN)
y = np.zeros(LEN)

plt.plot(y_actual, MSE(y, y_actual), label="Mean Squared Error")
plt.plot(y_actual, MAE(y, y_actual), label="Mean Absolute Error")
plt.plot(y_actual, SMAE(y, y_actual), label="Huber Loss")
plt.plot(y_actual, log_cosh(y, y_actual), label="Log Cosh")
plt.legend()
plt.show()
```
